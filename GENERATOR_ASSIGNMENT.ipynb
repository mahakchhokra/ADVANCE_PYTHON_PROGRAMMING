
# Write a generator function using yield that returns multiple values and print them using next()



```python

def my_generator():
    yield "First Value"
    yield "Second Value"
    yield "Third Value"

gen = my_generator()

print(next(gen))
print(next(gen))
print(next(gen))
```

# Write a program to iterate over a generator using a for loop and display all values


```python

def my_generator():
    yield "Apple"
    yield "Banana"
    yield "Cherry"
    yield "Date"

for fruit in my_generator():
    print(fruit)
```

# Write a generator expression to generate square values and print first few elements using next()


```python

squares = (x**2 for x in range(1, 11))

print(next(squares))
print(next(squares))
print(next(squares))
print(next(squares))
print(next(squares))
```

# Write a generator function to generate Fibonacci series and print the output as a list


```python

def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

fib_list = list(fibonacci(10))

print(fib_list)
```

# Write a program to demonstrate generator methods like next() and send() to control the flow of execution


```python

def number_generator():
    num = 0
    while True:

        new_value = yield num
        if new_value is not None:
            num = new_value
        else:
            num += 1

gen = number_generator()

print(next(gen))

print(next(gen))
print(next(gen))

print(gen.send(10))

print(next(gen))
print(next(gen))
```
