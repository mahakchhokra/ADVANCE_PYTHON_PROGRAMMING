```python
class Bird:
    def fly(self):
        print("Birds can fly.")

class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly.")

b= Bird()
p= Penguin()

b.fly()
p.fly()
```

```python


class Cat:
    def sound(self):
      return "Meow"

class Dog:
    def sound(self):
      return "Woof"

animals = [Cat(), Dog()]
for animal in animals:
    print(animal.sound())

```

```python
#2. Polymorphism with Inheritance
#A parent class method is overridden by child classes.


class Vehicle:
  def move(self):
    return "Moving"
class Car(Vehicle):
    def move(self):
      return "Car is driving"

class Boat(Vehicle):
    def move(self):
      return "Boat is sailing"

# Using Polymorphism
vehicles = [Car(), Boat()]
for v in vehicles:
    print(v.move())
```

```python
class Rectangle:
    def area(self, w, h): return w * h

class Circle:
    def area(self, r): return 3.14 * r * r

# Using Polymorphism
shapes = [Rectangle(), Circle()]
print(shapes[0].area(5, 10))  # Rectangle
print(shapes[1].area(7))      # Circle
```
