```python
class BankAccount:
    def __init__(self, balance, acc_no):
        self.__balance = balance
        self.acc_no = acc_no

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Rs.", amount, "credited successfully.")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount.")
        elif amount > self.__balance:
            print("Insufficient balance.")
        else:
            self.__balance -= amount
            print("Rs.", amount, "debited successfully.")

    def check_balance(self):
        return self.__balance


acc1 = BankAccount(10000, 23456)

acc1.deposit(2000)
acc1.withdraw(5000)

print("Current Balance:", acc1.check_balance())
```

```python
class Movie:
    def __init__(self, rating):
        self.__rating = rating

    def get_rating(self):
        return self.__rating

    def set_rating(self, rating):
        if 0 <= rating <= 10:
            self.__rating = rating
        else:
            print("Invalid rating. Must be between 1 and 10.")

m1 = Movie(8)

print("Rating:", m1.get_rating())

m1.set_rating(9)
print("Updated Rating:", m1.get_rating())

m1.set_rating(15)

```

```python
class Laptop:
    def __init__(self, price):
        self.__price = price

    def set_price(self, new_price):
        if new_price > 0:
            self.__price = new_price

    def get_price(self):
        return self.__price

l1 = Laptop(50000)

print("Price:", l1.get_price())   

l1.set_price(60000)
print("New price:",l1.get_price())   
```

```python
class Car:
    def __init__(self, speed=0):
        self.__speed = speed

    def increase_speed(self, amount):
        if amount > 0: self.__speed += amount

    def decrease_speed(self, amount):
        if 0 < amount <= self.__speed: self.__speed -= amount

    def get_speed(self):
        return self.__speed
c1 = Car()

c1.increase_speed(50)
print("Increase_speed:",c1.get_speed())   

c1.decrease_speed(20)
print("Decrease_speed:",c1.get_speed())   
```

```python
class BankAccount:
    def __init__(self, name, balance, acc_no):
        self.name = name
        self.__balance = balance
        self.acc_no = acc_no

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(self.name, "deposited Rs.", amount, "successfully.")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount, mode):
        if amount <= 0:
            print("Invalid withdrawal amount.")
        elif amount > self.__balance:
            print("Insufficient balance.")
        else:
            self.__balance -= amount
            print(self.name, "withdraw Rs.", amount, "successfully by", mode)

    def check_balance(self):
        return self.__balance


acc1 = BankAccount("Mahak", 10000, 23456)

acc1.deposit(2000)
acc1.withdraw(5000, "cheque")

print("Current Balance:", acc1.check_balance())
```
