# Create a simple decorator that adds extra print statements before and after a function execution


```python
def my_decorator(func):
    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")
    return wrapper


@my_decorator
def greet():
    print("Hello, World!")


greet()
```

# Create a decorator that wraps a function and prints a custom message before calling the original function


```python
def message_decorator(func):
    def wrapper():
        print("Starting the function...")
        func()
    return wrapper


@message_decorator
def welcome():
    print("Welcome to Python Decorators!")


welcome()
```

# Create a decorator that takes a function with two arguments and logs the function name along with its parameters


```python
def log_decorator(func):
    def wrapper(a, b):
        print(f"Function Name: {func.__name__}")
        print(f"Parameters: a = {a}, b = {b}")
        return func(a, b)
    return wrapper


@log_decorator
def add(x, y):
    return x + y


result = add(10, 20)
print("Result:", result)
```

# Create a decorator to measure execution time of a function that takes no arguments


```python
import time

def time_decorator(func):
    def wrapper():
        start_time = time.time()

        func()

        end_time = time.time()
        print(f"Execution Time: {end_time - start_time:.4f} seconds")

    return wrapper


@time_decorator
def example_function():
    print("Function is running...")
    time.sleep(2)


example_function()
```

# Create a decorator that checks if the input user is "admin" before allowing function execution


```python

def admin_required(func):
    def wrapper(user):
        if user == "admin":
            print("Access granted")
            func(user)
        else:
            print("Access denied")
    return wrapper


@admin_required
def dashboard(user):
    print(f"Welcome {user} to the admin dashboard")

dashboard("admin")
dashboard("guest")
```

# Create a decorator that caches results of a function with one integer argument to avoid repeated calculations


```python

def cache_decorator(func):
    cache = {}

    def wrapper(n):
        if n in cache:
            print(f"Returning cached result for {n}")
            return cache[n]

        result = func(n)
        cache[n] = result
        print(f"Calculated and cached result for {n}")
        return result

    return wrapper


@cache_decorator
def square(num):
    return num * num


# Testing
print(square(5))
print(square(5))
print(square(10))
print(square(10))
```

# Create a decorator that works with functions having arguments using *args and **kwargs and prints messages before and after execution


```python
def message_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before function execution")

        result = func(*args, **kwargs)

        print("After function execution")
        return result

    return wrapper


@message_decorator
def multiply(a, b):
    print(f"Multiplication: {a * b}")


@message_decorator
def greet(name="Guest"):
    print(f"Hello, {name}!")

multiply(4, 5)
greet(name="Mahak")
```

# Create a decorator that modifies the return value of a function by doubling the result


```python
def double_result(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result * 2
    return wrapper


@double_result
def add(a, b):
    return a + b


@double_result
def number():
    return 10

print(add(5, 3))
print(number())
```
