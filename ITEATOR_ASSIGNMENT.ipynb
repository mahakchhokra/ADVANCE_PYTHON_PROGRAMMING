 # Write a program to create an iterator from a list and print all elements using a while loop and next()


```python
# Create a list
my_list = [10, 20, 30, 40, 50]

# Create an iterator from the list
iterator = iter(my_list)

# Use while loop with next()
while True:
    try:
        # Get the next element
        element = next(iterator)
        print(element)
    except StopIteration:
        # Break when no more elements
        break
```

# Write a program to demonstrate what happens when next() is called more times than the number of elements in an iterator


```python

numbers = [1, 2, 3]

it = iter(numbers)

while True:
    try:
        element = next(it)
        print(element)
    except StopIteration:
        print("Iterator exhausted, stopping.")
        break
```

# Write a program to iterate over an iterator using a for loop and print all elements


```python

fruits = ["apple", "banana", "cherry", "date"]

iterator = iter(fruits)

for item in iterator:
    print(item)
```

# Write a program to create an iterator using range() and print values step by step using next()


```python

numbers = iter(range(1, 6))

print(next(numbers))  # Output: 1
print(next(numbers))  # Output: 2
print(next(numbers))  # Output: 3
print(next(numbers))  # Output: 4
print(next(numbers))  # Output: 5
```

# Write a program to manually traverse a list using iter() and next() and handle StopIteration using try-except


```python

my_list = [10, 20, 30, 40, 50]

iterator = iter(my_list)

while True:
    try:

        item = next(iterator)
        print(item)
    except StopIteration:
        print("End of list reached.")
        break
```
